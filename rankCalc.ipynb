```python
import numpy as np
import pandas as pd
```

```python
suit = ['c', 'd', 'h', 's'] # clubs, diamonds, hearts, spades
rank = ['2', '3', '4', '5', '6', '7', '8', '9', 'T', 'J', 'Q', 'K', 'A']
```

```python
class Card:
    def __init__(self, rank, suit):
        self.rank = rank
        self.suit = suit

    def __repr__(self):
        return f"({self.rank}, {self.suit})"
    
def generate_deck():
    deck = []
    for s in suit:
        for r in rank:
            deck.append(Card(r, s))
    return deck

assert len(generate_deck()) == 52
```
